```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
```

```python
mnist=tf.keras.datasets.fashion_mnist
(training_images,training_labels),(testing_images,testing_labels)=mnist.load_data()
```

```python
class_labels=['T-shirt/top','Trouser','Pullover','Dress','Coat','Sandal','Shirt','Sneaker','Bag','Ankel boot']
```

```python
plt.figure()
plt.imshow(training_images[0])
plt.colorbar()
plt.grid(False)
```

```python
training_images=training_images/255.0
testing_images=testing_images/255.0
```

```python
plt.figure(figsize=(10,10))
for i in range(25):
  plt.subplot(5,5,i+1)
  plt.xticks([])
  plt.yticks([])
  plt.grid(False)
  plt.imshow(training_images[i],cmap=plt.cm.binary)
  plt.xlabel(class_labels[training_labels[i]])
plt.show()
```

```python
model=keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(128,activation='relu'),
    keras.layers.Dense(10,activation='softmax')
])
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(training_images,training_labels,epochs=10,verbose=1,batch_size=128)
```

```python
test_loss,test_acc=model.evaluate(testing_images,testing_labels,verbose=2)
```

```python
prediction=model.predict(testing_images)
np.argmax(prediction[0])
```
